```python
import nltk
import re
```

```python
nltk.download('punkt')
```

```python
question = open('question_1.txt','r').readlines()
answer = open('answer_1.txt','r').readlines()
```

```python
for idx,i in enumerate(question):
    question[idx]=i.replace('\n','')

for idx,i in enumerate(answer):
    answer[idx]=i.replace('\n','')
```

```python
words_q = []
words_a = []
for i in question:
    words_q.extend(nltk.word_tokenize(i))
for i in answer:
    words_a.extend(nltk.word_tokenize(i))
```

```python
len(words_q)
```

```python
word2count = {}
for qs in words_q:
    if qs not in word2count:
        word2count[qs] = 1
    else:
        word2count[qs] += 1
        
word2count_a = {}
for qs in words_a:
    if qs not in word2count:
        word2count[qs] = 1
    else:
        word2count[qs] += 1
```

```python
len(word2count)
```

### Remove yang di bawah count 20 (kurang banyak dipakai)

```python
threshold=30
questionswords2int = {}
word_number = 0
for word, count in word2count.items():
    if count >= threshold:
        questionswords2int[word] = word_number
        word_number += 1
answerswords2int = {}
word_number = 0
for word, count in word2count.items():
    if count >= threshold:
        answerswords2int[word] = word_number
        word_number += 1
```

```python
len(questionswords2int),len(answerswords2int)
```

### Masukkan 4 token utama

- PAD : padding
- EOS : end of sentence
- OUT : yang tidak ditemukan
- SOS : end of sentence apa ya?

```python
tokens = ['<PAD>','<EOS>','<OUT>','<SOS>']
for token in tokens:
    questionswords2int[token] = len(questionswords2int) + 1
for token in tokens:
    answerswords2int[token] = len(answerswords2int) + 1
```

```python
questionint2word = dict(zip(questionswords2int.values(),questionswords2int.keys()))
answerint2word = dict(zip(answerswords2int.values(),answerswords2int.keys()))
```

### Tambahkan EOS

```python
for idx, ls in enumerate(question):
    question[idx] = ls + " <EOS>"
for idx, ls in enumerate(answer):
    answer[idx] = ls + " <EOS>"
```

```python
question
```

### Konversi ke NUMERIK dan replace OUT

```python
trn_question = []
for idx,qst in enumerate(question):
    phr = qst.split()
    for idx, i in enumerate(phr):
        try:
            phr[idx] = questionswords2int[i]
        except:
            phr[idx] = questionswords2int['<OUT>']
    trn_question.append(phr)
    
trn_answer = []
for idx,answ in enumerate(answer):
    phr = answ.split()
    for idx, i in enumerate(phr):
        try:
            phr[idx] = answerswords2int[i]
        except:
            phr[idx] = answerswords2int['<OUT>']
    trn_answer.append(phr)
```

### Check

```python
for i in range(10):
    print(answer[i],trn_answer[i])
```

### Sortir dari panjang yang paling sedikit

```python
maxlen=25

sort_question_n = []
sort_answer_n = []
for i in range(1,maxlen):
    for idx, qst in enumerate(trn_question):
        if len(qst) == i:
            sort_question_n.append(trn_question[idx])
            sort_answer_n.append(trn_answer[idx])
```

### Litte check

```python
def konv_ques(slist):
    rslt = ""
    for i in slist:
        rslt = rslt + questionint2word[i] + " "
    return print(rslt)
```

```python
def konv_ans(slist):
    rslt = ""
    for i in slist:
        rslt = rslt + answerint2word[i] + " "
    return print(rslt)
```
